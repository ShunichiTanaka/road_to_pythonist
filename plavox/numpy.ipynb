## 乱数生成

### 一様乱数
(0.0 〜 1.0)の間のランダムな値

```
# 一様乱数を1件出力する
np.random.rand()
```

```
# 一様乱数を3件出力する
np.random.rand(3)
```

```
np.random.rand(3, 4)
```

### 正規乱数
正規分布に従う乱数  
引数：平均、標準偏差、出力する件数  
デフォルト：0.0、1.0、1

```
# 標準正規乱数 (平均:0.0, 標準偏差:1.0) に従う乱数を 1 件出力
np.random.normal()
```

```
# 平均:50, 標準偏差:10 の正規分布に従う乱数を 1 件出力
np.random.normal(50, 10)
```

```
# 平均:50, 標準偏差:10 の正規分布に従う乱数を 3  x 4 の行列で出力する
np.random.normal(50, 10, (3, 4))
```

## 金融・財務計算

### 将来価値
numpy.fv()：ローンや投資の将来価値を求める

numpy.fv(rate, nper, pmt, pv, when='end')
- rate (必須): 利率
- nper (必須): 複利計算期間数 (支払回数)
- pmt (必須): 毎回の支払額
- pv (必須): 現在価値 (返済の場合: 借入金額, 貯蓄の場合: 積立済金額)
- when: 支払期日、いつ支払いが行われるか。 (end/0: 各期の期末, start/1: 各期の機種)

```
# 年利5%で10年間、毎月300ドル払い、現在価値=100ドルにおける将来価値
np.fv(0.5/12, 10*12, -300, -300)
```

### 利息を考慮したローンの月々の支払額(元金部分＋利息部分): PMT
np.pmt(rate, nper, pv, fv=0, when='end')

- rate (必須): 利率
- nper (必須): 複利計算期間数 (支払回数)
- pv (必須): 現在価値 (返済の場合: 借入金額, 貯蓄の場合: 積立済金額)
- fv : 将来価値 (返済の場合: 0, 貯蓄の場合: 積立目標額)
- when: 支払期日、いつ支払いが行われるか。 (end/0: 各期の期末, start/1: 各期の機種)

```
# 年利 0.8% の利率で 35 年間、3000 万円のローンを支払うのに必要な毎月の費用
np.pmt(0.008/12, 12*35, -3000*10000)
```
